# Merge 8-K Sample Data with CIK to CUSIP Data

```python
cols8k = ['CIK','Company Name','Form Type','Date Filed','Year','Quarter', 'Filename']
randsamp8k = pd.read_csv('C:\\Users\\whill\\OneDrive\\Desktop\\MGT of Fin Institutions\\Hills_A12\\8-Krandomsample.csv', usecols=(cols8k))
randsamp8k.head()
```

```python
randsamp8k.describe()
```

```python
cikcols = ['cusip','cik']
cikdata = pd.read_csv('C:\\Users\\whill\\OneDrive\\Desktop\\MGT of Fin Institutions\\Hills_A12\\cik_to_cusip.csv', usecols=(cikcols))
#used R and haven package to get this dataset from sas file 
cikdata.rename(columns ={'cik': 'CIK'}, inplace =True)
cikdata['cusip'] = cikdata['cusip'].str[0:6]
cikdata.head()
```

```python
converted8ks = pd.merge(cikdata, randsamp8k, how = 'inner', on=['CIK'])
converted8ks.head()
```

```python
converted8ks.describe()
```

```python
converted8ks.to_csv('C:\\Users\\whill\\OneDrive\\Desktop\\MGT of Fin Institutions\\Hills_A12\\8-K_CIKCUSIP.csv')
```

# Read in DSF Data merge with 8-K

```python
dsfdata = pd.read_csv('C:\\Users\\whill\\OneDrive\\Desktop\\MGT of Fin Institutions\\Hills_A10\\dsf_new.csv')
dsfdata['CUSIP'] = dsfdata['CUSIP'].str[0:6]
dsfdata.rename(columns = {'CUSIP': 'cusip'}, inplace = True)
dsfdata['year'] = pd.to_datetime(dsfdata['DATE'], format = '%Y%m%d').dt.year
dsfdata['RET'] = pd.to_numeric(dsfdata['RET'], errors='coerce')
dsfdata = dsfdata[(dsfdata['year'] >= 1970) & (dsfdata['year'] <= 2020)]
dsfdata['SHROUT'] = dsfdata['SHROUT']*1000
```

```python
converted8ks.rename(columns ={'Year': 'year'}, inplace =True)
fulldata = pd.merge(converted8ks,dsfdata, how = 'inner', left_on =['cusip', 'year'], right_on =['cusip', 'year'])
```

```python
fulldata.to_csv('C:\\Users\\whill\\OneDrive\\Desktop\\MGT of Fin Institutions\\Hills_A12\\fulldataA12.csv')
```
